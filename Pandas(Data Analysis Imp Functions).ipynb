# Data Wrangling :-
     
     Data wrangling is the process of removing errors and combining complex data sets to make them more 
    accessible and easier to analyze.
    A data wrangling process, also known as a data munging process, consists of reorganizing, transforming and mapping data
    from one "raw" form into another in order to make it more usable and valuable for a variety of downstream uses including 
    analytics.

The most commonly used examples of data wrangling are for:

-Merging several data sources into one data-set for analysis\
-Identifying gaps or empty cells in data and either filling or removing them\
-Deleting irrelevant or unnecessary data \
-Identifying severe outliers in data and either explaining the inconsistencies or deleting them to facilitate analysis   \
-Businesses also use data wrangling tools \

# Data Normalization :
    consists of transforming complex numeric columns to a standard scale.
    Method of data normalization: 
        1.simple feature scaling : Xnew=Xold/Xmax (range between o to 1)
        2. min- max scaling :   xnew = Xold-Xmin/ Xmax-Xmin (range between o to 1)
        3.z-score scaling :     Xnew=Xold-mean/std .......... std=Standard deviation (range between -1 to 1 or -3 to 3)


```python
df.head(2)
```

```python
df['DailyRateScale']=df['DailyRate']/df['DailyRate'].max()  #simple feature Scaling 
df[['DailyRate','DailyRateScale']].head()
```

```python
df['DailyRateScale2']=(df['DailyRate']-df['DailyRate'].min())/(df['DailyRate'].max()-df['DailyRate'].min()) #min-max scaling
df[['DailyRate','DailyRateScale2']].head(3)
```

```python
#z-score scaling
df['DailyRateScale3']=(df['DailyRate'] - df['DailyRate'].mean()) / df['DailyRate'].std()
df[['DailyRate','DailyRateScale3']].head()
```

# Drop Column

```python
df.drop(['DailyRateScale','DailyRateScale2','DailyRateScale3'] , axis=1, inplace=True)
df.columns
```

# Binning : 
    Process of converting Numerical data into categorical bins 
    Ex. Salary into [low , high, medium ] bins
for binning we use linspace() of numpy :\
syntax : numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

cut() to segment and sort the data values into bins .

```python
x = min(df['DistanceFromHome'])
y = max(df['DistanceFromHome'])
print(x)
print(y)
```

```python
bins = np.linspace(x,y,4)
group_name = ['Short','Medium','Long']
df['DistanceFromHomeBins'] = pd.cut(df['DistanceFromHome'], bins, labels=group_name,include_lowest=True)
bins
```

# Classify BusinessTravel based on TotalWorkingYears(low ,medium,high)

```python
x=min(df['TotalWorkingYears'])
y=max(df['TotalWorkingYears'])
bins=np.linspace(x,y,4)
bins
```

```python
group_name=['low','Medium','high']
df['TotalWorkingYearRATING']=pd.cut(df['TotalWorkingYears'], bins, labels= group_name,include_lowest=True)
df[['TotalWorkingYears','TotalWorkingYearRATING']].head()
```

```python
len(df['TotalWorkingYearRATING']=='low')
```

```python
def year_function(a):
    if a> 25:
        return "High"
    if a> 15:
        return "Medium"
    else:
        return "Low"
df['TWYRating']=df['TotalWorkingYears'].apply(year_function)
df[['TWYRating','TotalWorkingYears']].head()
```

```python
len(df.TWYRating=='Low')
```

# subplot() :you can draw multiple plots in one figure

```python
plt.figure(figsize=(5,5))
plt.subplot(2, 1, 1) 
sns.countplot(x='Gender',hue='TWYRating',data=df)
plt.title("TWYRating")

plt.subplot(2, 1, 2)
sns.countplot(x='Gender',hue='DistanceFromHomeBins',data=df)
plt.legend(loc='upper right',title='Distance')
plt.title("DistanceFromHomeBins")

plt.tight_layout()  #automatically adjusts subplots 
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(2,2,1) #the figure has 2 row, 2 columns, and this plot is the 1st plot.
plt.title('Department')
sns.countplot(x='Gender',hue='Department',data=df)

plt.subplot(2,2,2) #the figure has 2 row, 2 columns, and this plot is the 2nd plot
plt.title('EducationField')
sns.countplot(x='Gender',hue='EducationField',data=df)

plt.subplot(2,2,3)#the figure has 2 row, 2 columns, and this plot is the 3rd plot.
plt.title('BusinessTravel')
sns.countplot(x='Gender',hue='BusinessTravel',data=df)

plt.subplot(2,2,4)#the figure has 2 row, 2 columns, and this plot is the 4th plot.
plt.title('TWYRating')
sns.countplot(x='Gender',hue='TWYRating',data=df)

plt.tight_layout()
plt.show()
```
